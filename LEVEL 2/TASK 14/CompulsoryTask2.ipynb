```python
import pandas as pd
import numpy as numpy
from scipy import stats
from mlxtend.preprocessing import minmax_scaling
import seaborn as sns
import matplotlib.pyplot as plt
countries = pd.read_csv("countries.csv"
                              ,encoding = 'unicode_escape'
)
```

```python


#scaling 'EG.ELC.ACCS.ZS Column and visualizing the scaled data

original = countries['EG.ELC.ACCS.ZS']
scaled_countries = minmax_scaling(original, columns=[0])

plt.subplot(1,2,1)
plt.hist(original)

plt.subplot(1,2,2)
plt.hist(scaled_countries)




 
```

```python
#Standardizing the 'SP.DYN.CBRT.IN' column
sp = countries['SP.DYN.CBRT.IN']

sp_standardized = stats.boxcox(sp)[0]

plt.figure(figsize=(8,8))
plt.subplot(121)
plt.hist(sp)

plt.subplot(122)
plt.hist(sp_standardized)
```
