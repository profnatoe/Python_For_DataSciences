```python
#it is evident from our dataset that most of our columns are taking in numbers, the only column that takes in Strings is the Country and region. this cleanig this should be easy should there be a need.

#Cleaning the Country  and Region Column for 2015 dataset
#OVER HERE WE LOWER AND STRIP EXTRA WHITE SPACE TO PREPARE THE COLUMN FOR CLEANSING
worldhappiness2015['Region'] = worldhappiness2015['Region'].str.lower()
worldhappiness2015['Region'] = worldhappiness2015['Region'].str.strip()
worldhappiness2015['Region'].sort_values().unique()

worldhappiness2015['Country'] = worldhappiness2015['Country'].str.lower()
worldhappiness2015['Country'] = worldhappiness2015['Country'].str.strip()
worldhappiness2015['Country'].sort_values().unique()
```
