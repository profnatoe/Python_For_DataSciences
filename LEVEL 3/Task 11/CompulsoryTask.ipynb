```python
import numpy as np
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
%matplotlib inline

from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import accuracy_score, silhouette_score
```

```python
iris = datasets.load_iris()
iris_data = pd.DataFrame(iris.data)
iris_data.columns = iris.feature_names
iris_data['Type']=iris.target
iris_data
```

```python
iris_X = iris_data.iloc[:, [0,1]].values
iris_X
```

We are going to plot the dendrogram using the following :

1) method = complete


2) metric = euclidean

after we will use average method and the same metrics to plot the dendrograms


```python
z = linkage(iris_X, method = 'complete', metric='euclidean')
plt.figure(figsize=(20,7))

#plotting the dendrogram

den = dendrogram(z)
plt.title('Dendrogram for the clustering of the iris dataset')
plt.xlabel('Type')
plt.ylabel('Eucledian distance in the space with other variables')
```

```python
z = linkage(iris_X, method = 'complete', metric='cosine')
plt.figure(figsize=(20,7))

#plotting the dendrogram

den = dendrogram(z)
plt.title('Dendrogram for the clustering of the iris dataset')
plt.xlabel('Type')
plt.ylabel('Eucledian distance in the space with other variables')
```

We switch our method to average and plot the dendrograms


```python
z = linkage(iris_X, method = 'average', metric='euclidean')
plt.figure(figsize=(20,7))

#plotting the dendrogram

den = dendrogram(z)
plt.title('Dendrogram for the clustering of the iris dataset')
plt.xlabel('Type')
plt.ylabel('Eucledian distance in the space with other variables')
```

```python
z = linkage(iris_X, method = 'average', metric='cosine')
plt.figure(figsize=(20,7))

#plotting the dendrogram

den = dendrogram(z)
plt.title('Dendrogram for the clustering of the iris dataset')
plt.xlabel('Type')
plt.ylabel('Eucledian distance in the space with other variables')
```



We now do agglomerative clustering and we will choose the first dendrogram using method = complete and metric = euclidean

```python

complete = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')

complete_pred = complete.fit_predict(iris_X)

complete_pred
```

```python
ax = plt.scatter(iris_X[:,0], iris_X[:,1], c=complete.labels_)

score = round(silhouette_score(iris_X, complete.labels_, metric='euclidean'),2)
plt.title('silhouette={}'.format(score), loc='right', fontdict={'fontsize': 16}, pad=-14)
```

# the silhouette is 0.2 and as such it tells us that the clusters are overlapping. they are not dense and nicely separated. this is evident from the above scatter plot.

since it is positive as we know silhouette is between -1 and 1, this tells us that the data belonging to the clusters is correct
